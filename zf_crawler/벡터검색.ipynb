```python
import os
from dotenv import load_dotenv

# 환경변수 로드
load_dotenv()
```

```python
from src.database.repository import (
    AnncQrRepository
)
from src.embedder import embed_text

qr_repo = AnncQrRepository()
```

```python
def search_and_display(query: str, top_k: int = 5):
    """검색 실행 및 결과 표시"""
    print(f"\n{'='*60}")
    print(f"쿼리: {query}")
    print(f"{'='*60}")
    
    query_emb = embed_text(query)
    results = qr_repo.hybrid_search(query, query_emb, top_k=top_k)
    
    if not results:
        print("검색 결과 없음")
        return
    
    for i, row in enumerate(results):
        # row는 dict 형태
        chunk_id = row["chunk_id"]
        chunk_text = row["chunk_text"]
        element_type = row["chunk_type"]
        # page_num / table_context 컬럼을 sql_final에 포함했는지에 따라 사용
        page_num = row.get("page_num")
        table_context = row.get("table_context")
        annc_id = row["annc_id"]
        annc_title = row['annc_title']

        print(f"\n[{i+1}] 문서: {annc_title}\n, ID: {chunk_id}, 타입: {element_type}, 페이지: {page_num}")
        if table_context:
            print(f"    컨텍스트: {table_context}")
        print(f"    내용: {chunk_text[:200]}...")
```

```python
# 검색 테스트
search_and_display("임대보증금은 얼마인가요?")
```
